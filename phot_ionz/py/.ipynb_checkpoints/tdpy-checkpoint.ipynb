```python
def plot_mcmc(listsamp, suptitle='', strgpara=None, figsize=(6,6), nbins=100):
    
    nswep = listsamp.shape[0]
    npara = listsamp.shape[1]
    
    if strgpara == None:
        strgpara = [''] * npara
    
    fig, ax = plt.subplots(npara, npara, figsize=figsize)
    fig.suptitle(suptitle)
    for a, axrow in range(npara):
        for b, ax in range(npara):
            if a < b:
                ax.axis('off')
                continue
            ax.hist2d(listsamp[:,a], listsamp[:,b], nbins)
            ax.xlabel(strgpara[a])
            ax.ylabel(strgpara[b])
        ax.hist(listsamp[a], nbins)
        ax.xlabel(strgpara[a])
    plt.show()
```

```python
def adjust_axes(ax, xres=None, yres=None):

    if xres:
        start, stop = ax.get_xlim()
        ticks = np.arange(start, stop + xres, xres)
        ax.set_xticks(ticks)
    if yres:
        start, stop = ax.get_ylim()
        ticks = np.arange(start, stop + yres, yres)
        ax.set_yticks(ticks)
```

```python
def sample(pdfvar, arrvar, nsam, axis=0, getcdf=False, binned=False, rej=False):
    
    if rej:
        pdfvar -= np.amin(pdfvar, axis=axis)
        pdfvar /= np.amax(pdfvar, axis=axis)
        nsam_ = nsam
        samvar = []
        while nsam_ > 0:
            if not binned:
                rands = np.random.rand(nsam_ * 2)
                maxvar = max(arrvar)
                minvar = min(arrvar)
                samvar_ = rands[0:nsam_] * (maxvar - minvar) + minvar
                jsam = np.where(rands[nsam_:nsam_*2] - interpolate.interp1d(arrvar, pdfvar)(samvar_) < 0)[0]
                samvar.extend(samvar_[jsam])
                nsam_ -= jsam.size
            else:
                rands = np.random.rand(nsam_)
                samjvar = np.random.random_integers(0, arrvar.size-1, size=nsam_)
                jsam = np.where(rands < pdfvar[samjvar])[0]
                jvar = samjvar[jsam]
                samvar.extend(arrvar[jvar])
                nsam_ -= jvar.size
                
            
            #print 'nsam_: ', nsam_
            
        samvar = np.array(samvar)
    else:
        pdfvar /= np.trapz(pdfvar, arrvar)
        cdfvar = integrate.cumtrapz(pdfvar, arrvar, axis=axis, initial=0.)
        rands = np.random.rand(nsam)
        if not binned:
            samvar = interpolate.interp1d(cdfvar, arrvar)(rands)
        else:
            #print cdfvar
            #print cdfvar[:,None] - rands[None,:]
            jvar = np.argmin(np.fabs(cdfvar[:,None] - rands[None,:]), axis=0)
            #print jvar
            samvar = arrvar[jvar]
            #print samvar
            
    if getcdf and (not rej):
        return samvar, cdfvar
    else:
        return samvar
```

```python
def get_atcr(sgnl, ndela=10):
    
    nsgnl = sgnl.size
    
    meansgnlsqrd = np.mean(sgnl)**2

    atcr = np.empty(ndela)
    sgnllist = np.zeros((2, nsgnl))
    sgnllist[0,:] = sgnl
    for t in range(ndela):
        sgnllist[1,0:nsgnl-t] = sgnl[t:nsgnl]
        atcr[t] = np.mean(np.roll(sgnl, t) * sgnl) - meansgnlsqrd
        
    # normalize the autocorrelation
    vari = np.var(sgnl)
    atcr /= vari
         
    iact = 1. + 2. * np.sum(atcr[1:-1])
    return atcr, iact
```
